# Test data generator

This notebook creates testdata to evaluate the ellie RAG pipeline. The testdata are automatically generated through ragas https://docs.ragas.io/en/stable/.
ragas takes the  DnD source as input and generates question, answer context tripplets. These tripplets function as ground truth for the evaluation step.

```python
## Load Env
from dotenv import load_dotenv
load_dotenv()
```

## Load the Docs

```python
from langchain_community.document_loaders import DirectoryLoader
loader = DirectoryLoader("data")
documents = loader.load()

for document in documents:
    document.metadata['filename'] = document.metadata['source']
```

## Generate testset


```python

from ragas.testset.generator import TestsetGenerator
from ragas.testset.evolutions import simple, reasoning, multi_context
from langchain_openai import ChatOpenAI, OpenAIEmbeddings


import nest_asyncio
nest_asyncio.apply()

# generator with openai models
generator_llm = ChatOpenAI(model="gpt-3.5-turbo-16k")
critic_llm = ChatOpenAI(model="gpt-4")
embeddings = OpenAIEmbeddings()

generator = TestsetGenerator.from_langchain(
    generator_llm,
    critic_llm,
    embeddings,
)

# we run into timeouts if we generate 20 testdata from the whole rulebook, thats why we split into 2 runs

testset1 = generator.generate_with_langchain_docs(documents[:8], test_size=10, with_debugging_logs=True, distributions={simple: 1, reasoning: 0, multi_context: 0})
testset1_pandas = testset1.to_pandas()
testset1_pandas.to_csv("testset.csv", index=False)

testset2 = generator.generate_with_langchain_docs(documents[8:], test_size=10, with_debugging_logs=True, distributions={simple: 1, reasoning: 0, multi_context: 0})
testset2_pandas = testset2.to_pandas()
testset2_pandas.to_csv("testset2.csv", index=False)
```

```python
import pandas as pd

pd.concat([testset1_pandas, testset2_pandas]).to_csv("testset_combined.csv", index=False)
```
